## 1. File Operations - Reading

```python
import os

# Create a sample file for demonstration
sample_file = "sample.txt"
with open(sample_file, "w") as f:
    f.write("Line 1: Hello\n")
    f.write("Line 2: Python\n")
    f.write("Line 3: File I/O\n")

# Read entire file
with open(sample_file, "r") as f:
    content = f.read()
    print("Entire file:")
    print(content)

print("\n" + "="*40 + "\n")

# Read line by line
print("Line by line:")
with open(sample_file, "r") as f:
    for line_num, line in enumerate(f, 1):
        print(f"Line {line_num}: {line.rstrip()}")
```

## 2. File Operations - Writing

```python
# Write (overwrites existing content)
output_file = "output.txt"
with open(output_file, "w") as f:
    f.write("This is new content\n")
    f.write("It replaces old content\n")

# Append (adds to existing content)
with open(output_file, "a") as f:
    f.write("This is appended\n")

# Read and display
with open(output_file, "r") as f:
    print("Updated file:")
    print(f.read())

# Writing multiple lines
lines_file = "lines.txt"
lines = ["Python", "is", "awesome"]
with open(lines_file, "w") as f:
    f.writelines("\n".join(lines))

with open(lines_file, "r") as f:
    print("\nLines file:")
    print(f.read())
```

## 3. Working with CSV Files

```python
import csv

# Write CSV
csv_file = "data.csv"
data = [
    ["Name", "Age", "Score"],
    ["Alice", 25, 95],
    ["Bob", 30, 87],
    ["Charlie", 28, 92]
]

with open(csv_file, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(data)

# Read CSV
print("CSV Content:")
with open(csv_file, "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)

# Read CSV as dictionaries
print("\nCSV as dictionaries:")
with open(csv_file, "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(row)
```

## 4. JSON Files

```python
import json

# Write JSON
data = {
    "name": "Alice",
    "age": 30,
    "skills": ["Python", "ML", "Data"],
    "address": {
        "city": "NYC",
        "zip": "10001"
    }
}

json_file = "data.json"
with open(json_file, "w") as f:
    json.dump(data, f, indent=2)

# Read JSON
print("JSON Content:")
with open(json_file, "r") as f:
    loaded_data = json.load(f)
    print(json.dumps(loaded_data, indent=2))

# JSON string operations
json_string = json.dumps(data, indent=2)
print(f"\nJSON string:\n{json_string}")

parsed = json.loads(json_string)
print(f"\nParsed name: {parsed['name']}")
```

## 5. Error Handling

Gracefully handle exceptions.

```python
# Try-except-else-finally
def safe_file_read(filename):
    try:
        with open(filename, "r") as f:
            content = f.read()
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found")
    except IOError:
        print("Error: Could not read file")
    else:
        print(f"Successfully read {len(content)} characters")
    finally:
        print("File operation completed")

# Test with existing file
safe_file_read("sample.txt")

print("\n" + "="*40 + "\n")

# Test with non-existing file
safe_file_read("nonexistent.txt")
```

## 6. Custom Exceptions

```python
class InvalidAgeError(Exception):
    """Custom exception for invalid age"""
    pass

class InvalidNameError(Exception):
    """Custom exception for invalid name"""
    pass

def validate_person(name, age):
    if not isinstance(name, str) or not name:
        raise InvalidNameError("Name must be a non-empty string")
    if not isinstance(age, int) or age < 0 or age > 150:
        raise InvalidAgeError("Age must be between 0 and 150")
    return f"Valid: {name}, age {age}"

# Test valid
try:
    print(validate_person("Alice", 30))
except (InvalidNameError, InvalidAgeError) as e:
    print(f"Validation error: {e}")

# Test invalid
try:
    print(validate_person("Bob", 200))
except (InvalidNameError, InvalidAgeError) as e:
    print(f"Validation error: {e}")
```

## 7. Debugging Techniques

```python
# Print debugging
def buggy_calculation(x, y):
    print(f"DEBUG: x={x}, y={y}")
    result = x + y
    print(f"DEBUG: result={result}")
    return result

buggy_calculation(5, 3)

# Using assert
def divide(a, b):
    assert b != 0, "Divisor cannot be zero"
    return a / b

try:
    print(f"10 / 2 = {divide(10, 2)}")
    print(f"10 / 0 = {divide(10, 0)}")
except AssertionError as e:
    print(f"Assertion failed: {e}")

# Type checking
def process_data(data):
    print(f"Type of data: {type(data)}")
    print(f"Value: {data}")
    if isinstance(data, list):
        print(f"List with {len(data)} items")
    elif isinstance(data, dict):
        print(f"Dict with keys: {data.keys()}")

process_data([1, 2, 3])
print()
process_data({"a": 1, "b": 2})
```

## 8. Logging

```python
import logging

# Configure logging
logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s - %(levelname)s - %(message)s'
)

logger = logging.getLogger(__name__)

# Different log levels
logger.debug("This is a debug message")
logger.info("This is an info message")
logger.warning("This is a warning message")
logger.error("This is an error message")

# Logging with variables
user = "Alice"
action = "login"
logger.info(f"User {user} performed action: {action}")
```

## 9. Context Managers

```python
from contextlib import contextmanager
import time

# Built-in context manager (with statement)
print("Using with statement for file:")
with open("context_test.txt", "w") as f:
    f.write("Auto-closed after use")
# File is automatically closed here

# Custom context manager
@contextmanager
def timer_context(name):
    print(f"Starting {name}...")
    start = time.time()
    try:
        yield
    finally:
        elapsed = time.time() - start
        print(f"Finished {name} in {elapsed:.4f} seconds")

# Using custom context manager
with timer_context("task"):
    time.sleep(0.5)
    print("Task is running...")
```

## 10. File System Operations
