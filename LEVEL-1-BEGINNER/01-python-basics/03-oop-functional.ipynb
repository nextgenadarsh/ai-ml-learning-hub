## 1. Classes and Objects

Building blocks of OOP.

```python
# Basic class definition
class Person:
    def __init__(self, name: str, age: int) -> None:
        """Constructor - initialize object"""
        self.name = name
        self.age = age
    
    def introduce(self) -> str:
        """Instance method"""
        return f"Hi, I'm {self.name} and I'm {self.age} years old."
    
    def celebrate_birthday(self) -> None:
        """Modify instance state"""
        self.age += 1
        print(f"{self.name} is now {self.age} years old!")

# Creating objects (instances)
person1 = Person("Alice", 30)
person2 = Person("Bob", 25)

print(person1.introduce())
print(person2.introduce())

person1.celebrate_birthday()
```

### Class Methods and Static Methods

```python
class Dog:
    species = "Canis familiaris"  # Class variable
    
    def __init__(self, name: str) -> None:
        self.name = name
    
    def bark(self) -> str:
        """Instance method"""
        return f"{self.name} says: Woof!"
    
    @classmethod
    def get_species(cls) -> str:
        """Class method - works with class, not instance"""
        return cls.species
    
    @staticmethod
    def is_good_boy() -> bool:
        """Static method - no access to instance or class"""
        return True

# Using methods
dog = Dog("Buddy")
print(dog.bark())
print(f"Species: {Dog.get_species()}")
print(f"Is good boy: {Dog.is_good_boy()}")
```

## 2. Inheritance

Creating new classes from existing ones.

```python
# Parent class
class Animal:
    def __init__(self, name: str) -> None:
        self.name = name
    
    def make_sound(self) -> str:
        return "Some generic sound"

# Child classes (subclasses)
class Cat(Animal):
    def make_sound(self) -> str:
        return f"{self.name} says: Meow!"

class Dog(Animal):
    def make_sound(self) -> str:
        return f"{self.name} says: Woof!"
    
    def fetch(self, item: str) -> str:
        return f"{self.name} fetched the {item}!"

# Using inheritance
cat = Cat("Whiskers")
dog = Dog("Buddy")

print(cat.make_sound())
print(dog.make_sound())
print(dog.fetch("ball"))

# Polymorphism - same method, different behavior
animals = [cat, dog]
for animal in animals:
    print(animal.make_sound())
```

## 3. Encapsulation

Protecting internal data.

```python
class BankAccount:
    def __init__(self, owner: str, balance: float = 0) -> None:
        self.owner = owner
        self._balance = balance  # Protected (convention: single underscore)
        self.__pin = 1234  # Private (double underscore)
    
    def deposit(self, amount: float) -> None:
        """Public method"""
        if amount > 0:
            self._balance += amount
            print(f"Deposited {amount}. New balance: {self._balance}")
    
    def withdraw(self, amount: float, pin: int) -> bool:
        """Verify PIN before withdrawing"""
        if self._verify_pin(pin):
            if amount <= self._balance:
                self._balance -= amount
                print(f"Withdrew {amount}. New balance: {self._balance}")
                return True
        return False
    
    def _verify_pin(self, pin: int) -> bool:
        """Protected method (intended for internal use)"""
        return pin == self.__pin
    
    @property
    def balance(self) -> float:
        """Property - looks like attribute but calls method"""
        return self._balance

account = BankAccount("Alice", 1000)
account.deposit(500)
account.withdraw(300, 1234)
print(f"Balance: {account.balance}")
```

## 4. Lambda Functions

Small anonymous functions.

```python
# Lambda syntax: lambda arguments: expression

# Simple lambda
add = lambda x, y: x + y
print(f"5 + 3 = {add(5, 3)}")

# Lambda with single argument
square = lambda x: x ** 2
print(f"Square of 7 = {square(7)}")

# Conditional lambda
is_even = lambda x: "even" if x % 2 == 0 else "odd"
print(f"10 is {is_even(10)}")
print(f"7 is {is_even(7)}")

# Lambda with no arguments
get_message = lambda: "Hello, World!"
print(get_message())
```

## 5. Map, Filter, and Reduce

Functional programming tools.

```python
# Map - apply function to all items
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, numbers))
print(f"Squared: {squared}")

# Filter - keep items that pass test
numbers = list(range(1, 11))
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(f"Even numbers: {evens}")

# Reduce - cumulative operation
from functools import reduce
numbers = [1, 2, 3, 4, 5]
product = reduce(lambda x, y: x * y, numbers)
print(f"Product: {product}")

# Combined example
data = list(range(1, 6))
result = reduce(lambda x, y: x + y, filter(lambda x: x % 2 == 1, data))
print(f"Sum of odd numbers: {result}")
```

## 6. Decorators

Functions that modify other functions.

```python
import time

# Simple decorator
def my_decorator(func):
    def wrapper():
        print("Something before the function")
        func()
        print("Something after the function")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()

print("\n" + "="*40 + "\n")

# Decorator with arguments
def repeat(times):
    def decorator(func):
        def wrapper():
            for _ in range(times):
                func()
        return wrapper
    return decorator

@repeat(3)
def greet():
    print("Hello!")

greet()

print("\n" + "="*40 + "\n")

# Timing decorator
def timer(func):
    def wrapper():
        start = time.time()
        result = func()
        end = time.time()
        print(f"Execution time: {end - start:.4f} seconds")
        return result
    return wrapper

@timer
def slow_function():
    time.sleep(0.5)
    print("Function completed")

slow_function()
```

## 7. ML-Relevant Class Example

A practical example for machine learning.

```python
from typing import List

class SimpleModel:
    """A basic ML model class"""
    
    def __init__(self, name: str) -> None:
        self.name = name
        self._is_trained = False
        self._weights = None
    
    def fit(self, X: List[List[float]], y: List[float]) -> None:
        """Train the model"""
        print(f"Training {self.name}...")
        self._weights = [0.1, 0.2]  # Simplified
        self._is_trained = True
    
    def predict(self, X: List[List[float]]) -> List[float]:
        """Make predictions"""
        if not self._is_trained:
            raise ValueError("Model not trained yet!")
        return [sum(x[i] * w for i, w in enumerate(self._weights)) for x in X]
    
    def score(self) -> float:
        """Get model accuracy"""
        return 0.95 if self._is_trained else 0.0

# Using the model
model = SimpleModel("LinearRegressor")
X_train = [[1, 2], [2, 3], [3, 4]]
y_train = [3, 5, 7]

model.fit(X_train, y_train)
predictions = model.predict(X_train)
print(f"Predictions: {predictions}")
print(f"Score: {model.score()}")
```
