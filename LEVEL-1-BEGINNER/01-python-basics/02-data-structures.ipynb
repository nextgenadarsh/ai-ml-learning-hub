## 1. Lists

Ordered, mutable collections of items.

```python
# Creating lists
empty_list = []
numbers = [1, 2, 3, 4, 5]
mixed = [1, "hello", 3.14, True, None]

print(f"Empty list: {empty_list}")
print(f"Numbers: {numbers}")
print(f"Mixed types: {mixed}")

# Indexing and slicing
print(f"First element: {numbers[0]}")
print(f"Last element: {numbers[-1]}")
print(f"Slice [1:3]: {numbers[1:3]}")
print(f"Slice [::2]: {numbers[::2]}")

# Length
print(f"Length: {len(numbers)}")
```

### List Methods

```python
fruits = ['apple', 'banana', 'cherry']

# Append
fruits.append('date')
print(f"After append: {fruits}")

# Insert
fruits.insert(1, 'blueberry')
print(f"After insert: {fruits}")

# Remove
fruits.remove('blueberry')
print(f"After remove: {fruits}")

# Pop
last = fruits.pop()
print(f"Popped: {last}, List: {fruits}")

# Sort
numbers = [3, 1, 4, 1, 5]
numbers.sort()
print(f"Sorted: {numbers}")

# Reverse
numbers.reverse()
print(f"Reversed: {numbers}")

# Index and count
fruits = ['apple', 'banana', 'apple']
print(f"Index of 'banana': {fruits.index('banana')}")
print(f"Count of 'apple': {fruits.count('apple')}")
```

## 2. Tuples

Ordered, immutable collections.

```python
# Creating tuples
empty_tuple = ()
single_tuple = (1,)  # Note the comma
tuple_1 = (1, 2, 3)
tuple_2 = 1, 2, 3  # Parentheses optional
mixed_tuple = (1, "hello", 3.14)

print(f"Single tuple: {single_tuple}, Type: {type(single_tuple)}")
print(f"Tuple 1: {tuple_1}")
print(f"Tuple 2: {tuple_2}")

# Unpacking
a, b, c = tuple_1
print(f"Unpacked: a={a}, b={b}, c={c}")

# Immutability
try:
    tuple_1[0] = 10  # This will cause an error
except TypeError as e:
    print(f"Error (expected): {e}")

# Tuple methods
t = (1, 2, 3, 2, 1)
print(f"Index of 2: {t.index(2)}")
print(f"Count of 1: {t.count(1)}")
```

## 3. Dictionaries

Unordered, mutable collections of key-value pairs.

```python
# Creating dictionaries
empty_dict = {}
person = {'name': 'Alice', 'age': 30, 'city': 'NYC'}
scores = dict(math=95, science=88, english=92)

print(f"Empty dict: {empty_dict}")
print(f"Person: {person}")
print(f"Scores: {scores}")

# Accessing values
print(f"Name: {person['name']}")
print(f"Age: {person.get('age')}")
print(f"Country (not exists): {person.get('country', 'Unknown')}")

# Checking keys
print(f"'name' in person: {'name' in person}")
print(f"'country' in person: {'country' in person}")
```

### Dictionary Methods

```python
d = {'a': 1, 'b': 2, 'c': 3}

# Get keys, values, items
print(f"Keys: {d.keys()}")
print(f"Values: {d.values()}")
print(f"Items: {d.items()}")

# Add/update
d['d'] = 4
d.update({'e': 5, 'f': 6})
print(f"After adding: {d}")

# Remove
removed = d.pop('a')
print(f"Removed 'a': {removed}, Dict: {d}")

# Clear all
d_copy = d.copy()
d_copy.clear()
print(f"Cleared dict: {d_copy}")
```

## 4. Sets

Unordered collections of unique items.

```python
# Creating sets
set_1 = {1, 2, 3, 4}
set_2 = {3, 4, 5, 6}
empty_set = set()  # Not {}, that's a dict

print(f"Set 1: {set_1}")
print(f"Set 2: {set_2}")

# Removing duplicates
numbers = [1, 2, 2, 3, 3, 3]
unique = set(numbers)
print(f"Unique: {unique}")

# Set operations
print(f"Union: {set_1 | set_2}")
print(f"Intersection: {set_1 & set_2}")
print(f"Difference: {set_1 - set_2}")
print(f"Symmetric difference: {set_1 ^ set_2}")

# Membership
print(f"2 in set_1: {2 in set_1}")
print(f"10 in set_1: {10 in set_1}")
```

### Set Methods

```python
s = {1, 2, 3}

# Add and remove
s.add(4)
print(f"After add: {s}")

s.remove(2)
print(f"After remove: {s}")

s.discard(10)  # Doesn't error if not exists
print(f"After discard: {s}")

# Pop
item = s.pop()
print(f"Popped: {item}, Set: {s}")
```

## 5. List Comprehensions

Concise way to create lists.

```python
# Basic comprehension
squares = [x**2 for x in range(5)]
print(f"Squares: {squares}")

# With condition
even_squares = [x**2 for x in range(10) if x % 2 == 0]
print(f"Even squares: {even_squares}")

# Nested comprehension
matrix = [[i*j for j in range(3)] for i in range(3)]
for row in matrix:
    print(row)

# String comprehension
text = "hello"
uppercase = [char.upper() for char in text]
print(f"Uppercase: {uppercase}")

# Dictionary comprehension
squares_dict = {x: x**2 for x in range(5)}
print(f"Squares dict: {squares_dict}")

# Set comprehension
unique_lengths = {len(word) for word in ['apple', 'car', 'dog', 'cat']}
print(f"Unique lengths: {unique_lengths}")
```

## 6. Nested Data Structures

```python
# List of dictionaries (common in data processing)
students = [
    {'name': 'Alice', 'age': 20, 'gpa': 3.8},
    {'name': 'Bob', 'age': 21, 'gpa': 3.5},
    {'name': 'Charlie', 'age': 20, 'gpa': 3.9}
]

for student in students:
    print(f"{student['name']}: GPA {student['gpa']}")

# Dictionary with lists
data = {
    'scores': [95, 87, 92],
    'names': ['Alice', 'Bob', 'Charlie']
}

for i, (name, score) in enumerate(zip(data['names'], data['scores'])):
    print(f"{i}: {name} - {score}")

# Deeply nested
nested = {
    'level1': {
        'level2': {
            'level3': [1, 2, 3]
        }
    }
}

print(f"Deeply nested value: {nested['level1']['level2']['level3'][1]}")
```
