## Docker and Containerization

### Why Docker?
- **Consistency**: Works same way everywhere
- **Isolation**: Independent environments
- **Reproducibility**: Version all dependencies
- **Easy deployment**: Push and run

### Dockerfile Example
```dockerfile
FROM python:3.10-slim

WORKDIR /app
COPY requirements.txt .
RUN pip install -r requirements.txt

COPY . .

EXPOSE 8000
CMD ["python", "-m", "uvicorn", "app:app", "--host", "0.0.0.0"]
```

### Building and Pushing
```bash
docker build -t myregistry/mymodel:v1.0 .
docker push myregistry/mymodel:v1.0
docker run -p 8000:8000 myregistry/mymodel:v1.0
```

## Kubernetes Fundamentals

### Kubernetes Architecture
```
┌─────────────────────────────────────┐
│      Kubernetes Cluster             │
├─────────────────────────────────────┤
│  Master Node (Control Plane)        │
│  - API Server                       │
│  - Scheduler                        │
│  - Controller Manager               │
├─────────────────────────────────────┤
│  Worker Nodes                       │
│  ┌─────────────┐ ┌─────────────┐   │
│  │ Pod Pod Pod │ │ Pod Pod Pod │   │
│  └─────────────┘ └─────────────┘   │
└─────────────────────────────────────┘
```

### Core Concepts
1. **Pod**: Smallest deployable unit
   - Contains one or more containers
   - Shared networking

2. **Deployment**: Manage pod replicas
   - Desired state
   - Automatic rollout/rollback
   - Scaling

3. **Service**: Network abstraction
   - Load balancing
   - Service discovery
   - Stable IP

### Deployment YAML
```yaml
apiVersion: apps/v1
kind: Deployment
metadata:
  name: ml-model-api
spec:
  replicas: 3
  selector:
    matchLabels:
      app: ml-model
  template:
    metadata:
      labels:
        app: ml-model
    spec:
      containers:
      - name: model-container
        image: myregistry/mymodel:v1.0
        ports:
        - containerPort: 8000
        resources:
          requests:
            memory: "256Mi"
            cpu: "250m"
          limits:
            memory: "512Mi"
            cpu: "500m"
```

## Infrastructure as Code

### Why IaC?
- Reproducible environments
- Version control infrastructure
- Automated provisioning
- Disaster recovery

### Terraform Example
```hcl
# AWS resources
resource "aws_ecr_repository" "model" {
  name = "ml-model"
}

resource "aws_ecs_cluster" "ml" {
  name = "ml-cluster"
}

resource "aws_ecs_service" "model" {
  name            = "ml-model-service"
  cluster         = aws_ecs_cluster.ml.id
  task_definition = aws_ecs_task_definition.model.arn
  desired_count   = 3

  load_balancer {
    target_group_arn = aws_lb_target_group.model.arn
    container_name   = "model-container"
    container_port   = 8000
  }
}
```

### Tools
- **Terraform**: Cloud-agnostic
- **CloudFormation**: AWS-specific
- **Helm**: Kubernetes packages

## Cost Optimization

### Typical ML Costs Breakdown
- Compute (training, inference): 60-70%
- Storage (data, models): 15-20%
- Networking: 5-10%
- Tools/licenses: 5-10%

### Cost Reduction Strategies
1. **Compute**
   - Use spot/preemptible instances (60-90% cheaper)
   - Reserved instances for baseline
   - Auto-scaling: remove unused capacity
   - GPU sharing

2. **Storage**
   - Archive old data
   - Compression
   - Delete unused models

3. **Development**
   - Shared compute resources
   - Lower-cost instance types
   - Spot instances for experimentation
