```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

# Load dataset
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

print(f"Dataset shape: {df.shape}")
print(df.head())
```

## Univariate Analysis

Analyzing single variables to understand distributions and patterns.

```python
# Histogram and KDE
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Distribution of first feature
axes[0, 0].hist(df['mean radius'], bins=30, edgecolor='black', alpha=0.7)
axes[0, 0].set_title('Histogram: Mean Radius')
axes[0, 0].set_xlabel('Value')
axes[0, 0].set_ylabel('Frequency')

# KDE plot
df['mean radius'].plot(kind='kde', ax=axes[0, 1], title='KDE: Mean Radius')

# Box plot
df['mean radius'].plot(kind='box', ax=axes[1, 0], title='Box Plot: Mean Radius')

# Q-Q plot for normality
stats.probplot(df['mean radius'], dist='norm', plot=axes[1, 1])
axes[1, 1].set_title('Q-Q Plot: Mean Radius')

plt.tight_layout()
plt.show()
```

## Bivariate Analysis

Analyzing relationships between two variables.

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Scatter plot
axes[0].scatter(df['mean radius'], df['mean texture'], alpha=0.6)
axes[0].set_title('Scatter Plot')
axes[0].set_xlabel('Mean Radius')
axes[0].set_ylabel('Mean Texture')

# Scatter with target coloring
colors = df['target'].map({0: 'red', 1: 'blue'})
axes[1].scatter(df['mean radius'], df['mean texture'], c=colors, alpha=0.6)
axes[1].set_title('Scatter Plot (colored by target)')
axes[1].set_xlabel('Mean Radius')
axes[1].set_ylabel('Mean Texture')

# Regression plot
sns.regplot(x='mean radius', y='mean texture', data=df, ax=axes[2], scatter_kws={'alpha':0.5})
axes[2].set_title('Regression Plot')

plt.tight_layout()
plt.show()
```

```python
# Correlation Analysis
# Select numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns[:10]  # Select first 10 for clarity

# Correlation matrix
corr_matrix = df[numeric_cols].corr()

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0,
            square=True, linewidths=1)
plt.title('Correlation Matrix Heatmap')
plt.tight_layout()
plt.show()

# Find highly correlated features
print("Highly correlated pairs (correlation > 0.8):")
corr_pairs = []
for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        if abs(corr_matrix.iloc[i, j]) > 0.8:
            print(f"{corr_matrix.columns[i]} - {corr_matrix.columns[j]}: {corr_matrix.iloc[i, j]:.3f}")
```

```python
# Categorical EDA
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Target distribution
df['target'].value_counts().plot(kind='bar', ax=axes[0], color=['red', 'blue'])
axes[0].set_title('Target Distribution')
axes[0].set_xlabel('Class')
axes[0].set_ylabel('Count')

# Pie chart
df['target'].value_counts().plot(kind='pie', ax=axes[1], autopct='%1.1f%%')
axes[1].set_title('Target Distribution (%)')
axes[1].set_ylabel('')

plt.tight_layout()
plt.show()

# Statistics by target
print("\nMean values by target:")
print(df.groupby('target')[numeric_cols].mean())
```
