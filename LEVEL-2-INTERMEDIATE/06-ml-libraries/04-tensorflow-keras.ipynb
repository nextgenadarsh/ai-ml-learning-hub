```python
import tensorflow as tf
from tensorflow import keras
import numpy as np

print(f"TensorFlow version: {tf.__version__}")

# Tensors
scalar = tf.constant(5)
vector = tf.constant([1, 2, 3])
matrix = tf.constant([[1, 2], [3, 4]])

print("Scalar:", scalar)
print("Vector:", vector)
print("Matrix shape:", matrix.shape)

# Tensor operations
a = tf.constant([1, 2, 3], dtype=tf.float32)
b = tf.constant([4, 5, 6], dtype=tf.float32)

print("\nAddition:", tf.add(a, b))
print("Multiplication:", tf.multiply(a, b))
print("Matrix multiplication:", tf.matmul(tf.reshape(a, [3, 1]), tf.reshape(b, [1, 3])))
```

```python
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist

# Build a simple model
model = models.Sequential([
    layers.Dense(128, activation='relu', input_shape=(784,)),
    layers.Dropout(0.2),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(10, activation='softmax')
])

model.summary()

# Compile
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("Model compiled successfully!")
```

```python
# Load MNIST data
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize
x_train = x_train.reshape(-1, 784).astype('float32') / 255.0
x_test = x_test.reshape(-1, 784).astype('float32') / 255.0

print("Training data shape:", x_train.shape)
print("Test data shape:", x_test.shape)

# Train (reduced epochs for speed)
history = model.fit(
    x_train, y_train,
    epochs=5,
    batch_size=128,
    validation_split=0.1,
    verbose=1
)

# Evaluate
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
print(f"\nTest Accuracy: {test_acc:.4f}")
```

```python
# Convolutional Neural Network
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Reshape for CNN (28, 28, 1)
x_train = x_train.reshape(-1, 28, 28, 1).astype('float32') / 255.0
x_test = x_test.reshape(-1, 28, 28, 1).astype('float32') / 255.0

# Build CNN
cnn_model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(10, activation='softmax')
])

cnn_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("CNN Model Summary:")
cnn_model.summary()
```

```python
# Recurrent Neural Network
from tensorflow.keras.layers import LSTM, Embedding

# Build RNN for sequence
rnn_model = models.Sequential([
    layers.LSTM(64, activation='relu', input_shape=(10, 1), return_sequences=True),
    layers.LSTM(32, activation='relu'),
    layers.Dense(16, activation='relu'),
    layers.Dense(1)
])

rnn_model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)

print("RNN Model Summary:")
rnn_model.summary()

# Generate dummy data for demo
X_seq = np.random.rand(100, 10, 1)
y_seq = np.random.rand(100, 1)

# Train briefly
history = rnn_model.fit(X_seq, y_seq, epochs=3, verbose=0)
print("RNN training complete!")
```
