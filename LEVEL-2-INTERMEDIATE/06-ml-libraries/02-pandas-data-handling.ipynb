```python
import pandas as pd
import numpy as np

# Creating Series
s = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
print("Series:\n", s)
print("\nData type:", type(s))

# Creating DataFrames
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Salary': [50000, 60000, 75000]
}
df = pd.DataFrame(data)
print("\nDataFrame:\n", df)

# DataFrame info
print("\nShape:", df.shape)
print("Columns:", df.columns.tolist())
print("\nData types:\n", df.dtypes)
```

```python
# Data Selection
df = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [5, 6, 7, 8],
    'C': [9, 10, 11, 12]
})

# Select column
print("Column A:\n", df['A'])

# Select multiple columns
print("\nColumns A and C:\n", df[['A', 'C']])

# Select rows by position
print("\nFirst 2 rows:\n", df.iloc[0:2])

# Select by condition
print("\nRows where A > 2:\n", df[df['A'] > 2])

# loc vs iloc
print("\nUsing loc (label-based):\n", df.loc[0:1, 'A':'B'])
print("\nUsing iloc (position-based):\n", df.iloc[0:2, 0:2])
```

```python
# Data Operations
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice'],
    'Department': ['Sales', 'IT', 'Finance', 'Sales'],
    'Salary': [50000, 65000, 75000, 52000]
})

# Group by
grouped = df.groupby('Department')['Salary'].mean()
print("Average salary by department:\n", grouped)

# Aggregation
agg = df.groupby('Department').agg({
    'Salary': ['mean', 'sum', 'count']
})
print("\nAggregation:\n", agg)

# Sorting
print("\nSorted by salary:\n", df.sort_values('Salary', ascending=False))

# Value counts
print("\nValue counts:\n", df['Department'].value_counts())
```

```python
# Merging and Joining
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})

df2 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Salary': [50000, 65000, 75000]
})

# Merge (inner join by default)
merged = pd.merge(df1, df2, on='ID')
print("Merged (inner):\n", merged)

# Different join types
left_merge = pd.merge(df1, df2, on='ID', how='left')
print("\nLeft merge:\n", left_merge)

# Concatenation
concatenated = pd.concat([df1, df2], axis=1)
print("\nConcatenated (columns):\n", concatenated)
```

```python
# Time Series
dates = pd.date_range('2023-01-01', periods=6, freq='D')
ts = pd.Series(range(1, 7), index=dates)
print("Time Series:\n", ts)

# Resampling
ts_df = pd.DataFrame({'value': range(1, 31)}, index=pd.date_range('2023-01-01', periods=30))
weekly = ts_df.resample('W').sum()
print("\nWeekly sum:\n", weekly)

# Rolling window
rolling_mean = ts_df.rolling(window=7).mean()
print("\nRolling mean (window=7):\n", rolling_mean.head())
```

## Key Concepts

### DataFrame Structure
- Rows (index) and columns
- Can have multiple data types
- Labeled by default

### Common Operations
- **Selection**: [], .loc[], .iloc[]
- **Filtering**: Boolean indexing
- **Grouping**: groupby(), agg()
- **Joining**: merge(), concat(), join()
- **Cleaning**: dropna(), fillna(), replace()

### Best Practices
- Use vectorized operations
- Avoid loops
- Use apply() sparingly
- Cache aggregations
