````python
import numpy as np
import time

# Array Creation
arr1d = np.array([1, 2, 3, 4, 5])
arr2d = np.array([[1, 2, 3], [4, 5, 6]])
arr3d = np.zeros((3, 4, 5))

print("1D Array:", arr1d.shape, arr1d.dtype)
print("2D Array:", arr2d.shape)
print("3D Array:", arr3d.shape)

# Array Generation
zeros = np.zeros((3, 3))
ones = np.ones((2, 4))
random_arr = np.random.rand(3, 3)
range_arr = np.arange(0, 10, 2)
linspace_arr = np.linspace(0, 1, 5)

print("\nRandom array:", random_arr[0])
print("Range array:", range_arr)
print("Linspace array:", linspace_arr)
````

````python
# Element-wise Operations
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Power:", a ** 2)
print("Square root:", np.sqrt(a))

# Aggregate Functions
arr = np.array([1, 2, 3, 4, 5])
print("\nSum:", np.sum(arr))
print("Mean:", np.mean(arr))
print("Std Dev:", np.std(arr))
print("Min:", np.min(arr))
print("Max:", np.max(arr))
````

````python
# Linear Algebra
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Matrix multiplication
C = np.dot(A, B)
print("Matrix multiplication:\n", C)

# Matrix inverse
A_inv = np.linalg.inv(A)
print("\nInverse of A:\n", A_inv)

# Determinant
det_A = np.linalg.det(A)
print("\nDeterminant of A:", det_A)

# Eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)
print("\nEigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)

# Rank
rank = np.linalg.matrix_rank(A)
print("\nRank of A:", rank)
````

````python
# Broadcasting
a = np.array([[1, 2, 3],
               [4, 5, 6]])
b = np.array([1, 2, 3])

print("A shape:", a.shape)
print("B shape:", b.shape)
print("\nA + B (broadcasting):\n", a + b)

# Scalar broadcasting
scalar = 5
print("\nA * scalar:\n", a * scalar)

# Important: Shapes must be compatible
# (m, n) with (n,) works
# (m, 1) with (m, n) works
# (m, n) with (n,) broadcasts to (m, n)
````

````python
# Indexing and Slicing
arr = np.arange(20).reshape(4, 5)
print("Array:\n", arr)

# Basic indexing
print("\nElement at [0, 0]:", arr[0, 0])
print("Row 0:", arr[0, :])
print("Column 0:", arr[:, 0])

# Slicing
print("\nSubarray [0:2, 1:3]:\n", arr[0:2, 1:3])

# Boolean indexing
mask = arr > 10
print("\nElements > 10:", arr[mask])

# Fancy indexing
indices = [0, 2]
print("\nRows 0 and 2:\n", arr[indices])
````

````python
# Performance: NumPy vs Python Lists
n = 1000000

# Python list
list_a = list(range(n))
start = time.time()
result_list = [x * 2 for x in list_a]
list_time = time.time() - start

# NumPy array
arr_a = np.arange(n)
start = time.time()
result_arr = arr_a * 2
numpy_time = time.time() - start

print(f"List operation: {list_time:.6f} seconds")
print(f"NumPy operation: {numpy_time:.6f} seconds")
print(f"NumPy is {list_time/numpy_time:.1f}x faster!")
````

## Key Concepts

### Broadcasting Rules
1. If arrays have different dimensions, pad smaller shape with 1s
2. Check if dimensions are compatible:
   - Dimensions must be equal, or
   - One dimension must be 1
3. Iterate over broadcasted shapes

### Memory Efficiency
- NumPy arrays are homogeneous (same type)
- Fixed size (no dynamic resizing)
- Contiguous memory layout
- Vectorized operations

### Common Patterns
